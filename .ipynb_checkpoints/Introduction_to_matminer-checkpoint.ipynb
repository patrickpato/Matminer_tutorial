```python
#WORKING WITH MPDataRetrieval requires an API key. this can be easily generated from the materials project website
#https://materialsproject.org/. Ensure you sign in first then head over to the dashboard section
#here, we extract elemental compounds with the only properties being their density and formula
mpdr = MPDataRetrieval(api_key='XiwMxgnFdRz49DmU')
df = mpdr.get_dataframe(criteria={"nelements": 1}, properties=['density', 'pretty_formula'])#the pretty formula property is key. Any other name will provide a different output
print("There are {} entries on MP with 1 element".format(df['density'].count()))
```

```python
df.head()#provides a dataframe of all elemental compounds, i.e those containing only one element
```

```python
#investigating additional properties such as bandstructure, DOS, shear modulus etc
#here, compounds containing Pb and Te are parsed. 
#This filtering is based on their significant thermoelectric properties
multiple_properties_df = mpdr.get_dataframe(criteria={"elasticity": {"$exists": True}, 
                         "elasticity.warnings": [],
                         "elements": {"$all": ["Pb", "Te"]},
                         "e_above_hull": {"$lt": 1e-6}}, # to limit the number of hits for the sake of time
                        properties = ["elasticity.K_VRH", "elasticity.G_VRH", "pretty_formula", 
                                      "e_above_hull", "bandstructure", "dos"])
multiple_properties_df.head()
```

```python
len(multiple_properties_df)
```

```python
#vizualization of the bandstructure and DOS
%matplotlib inline
from pymatgen.electronic_structure.plotter import BSDOSPlotter

mpid = 'mp-20740'
idx = multiple_properties_df.index[multiple_properties_df.index==mpid][0]
plt = BSDOSPlotter().get_plot(bs=multiple_properties_df.loc[idx, 'bandstructure'], dos=multiple_properties_df.loc[idx, 'dos']);
plt.show()
```

```python
#Repeating the above provedure for CaTiO3 and searching for bandstructure and dos alone
CaTiO3_df = mpdr.get_dataframe(criteria={#"elasticity": {"$exists": True}, 
                         #"elasticity.warnings": [],
                         "elements": {"$all": ["Ca", "Ti", "O"]},
                         "e_above_hull": {"$lt": 1e-6}}, # to limit the number of hits for the sake of time
                        properties = [#"elasticity.K_VRH", "elasticity.G_VRH", "pretty_formula", 
                                      "e_above_hull", "bandstructure", "dos", "bandgap"])
CaTiO3_df.head()
```

it is important to take note of the fact that MPD lacks data for some queries as seen from the none values above

```python
#plotting the dos and bandstructure of the first entry
mpid = 'mp-4019'
idx = CaTiO3_df.index[CaTiO3_df.index==mpid][0]
plt = BSDOSPlotter().get_plot(bs=CaTiO3_df.loc[idx, 'bandstructure'], dos=CaTiO3_df.loc[idx, 'dos']);
plt.show()
```

The results above agree with most studies. For instance, it is known that Oxygen is dorminant in the valence band of the structure. Also, crystal field splitting is significantly evident in the conduction band of the compound above.
